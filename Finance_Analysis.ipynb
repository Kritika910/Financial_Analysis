# Preprocessing and V

```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('darkgrid')
```

```python
df = pd.read_csv("Finance_data.csv")
```

```python
print("\nFirst 5 rows:")
df.head()
```

```python
print("Data Overview:")
df.info()
```

```python
# Check for missing values
print("\nMissing Values:")
df.isnull().sum()
```

```python
#dropping the null values
df.dropna()
```

```python
# Summary statistics
print("\nSummary Statistics:")
df.describe(include='all')
```

```python
df.drop(['Mutual_Funds', 'Equity_Market', 'Debentures', 'Government_Bonds', 'PPF', 'Fixed_Deposits',"Gold" ], axis = 1, inplace = True)
```

```python
df = df.replace(r'^\s*$', np.nan, regex = True)
```

# DATA VISUALISATION

```python
sns.countplot(x='age', data = df, palette = "Set2", linewidth = 2,hue='gender',edgecolor = "black")
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1,2,1)
sns.countplot(x=df['gender'], hue = df['Investment_Avenues'], palette = 'summer', linewidth = 3, edgecolor = "white")
plt.title("INVESTMENT AVENUE")

plt.subplot(1,2,2)
sns.countplot(x= df['gender'], hue = df['Stock_Marktet'], palette = 'hot', linewidth = 3, edgecolor = 'white')
plt.title("STOCK MARKET")
plt.show()
```

```python
sns.countplot(x=df['Factor'],hue= df['gender'], palette = 'coolwarm', linewidth = 2, edgecolor = "black")
plt.show()
```

```python
plt.figure(figsize = (10,4))
sns.pointplot(x = "Purpose", y = "age", data = df, linestyles = "--", capsize = .3, color = "Blue")
plt.show()
```

```python
plt.figure(figsize= (18,6))
plt.subplot(1,2,1)
sns.countplot(hue = df['gender'], x = df['Duration'], palette = 'viridis', linewidth = 2, edgecolor = "black" )
plt.title("DURATION")

plt.subplot(1,2,2)
sns.countplot(hue = df["gender"], x = df["Invest_Monitor"], palette = "seismic", linewidth = 2, edgecolor ="black")
plt.title("INVESTMENT MONITORING")
plt.show()
```

```python
# Boxplot of age vs investment expectation
plt.figure(figsize=(12, 6))
sns.boxplot(x='Expect', y='age',hue='Expect', data=df, palette='Set3')
plt.title('Age vs Investment Expectation')
plt.show()
```
